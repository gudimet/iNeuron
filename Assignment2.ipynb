```python
def myfilter(string,filter_str):
    filter_out=[]
    for i in range(len(string)):
        for j in range(len(filter_str)):
             if string[i]==filter_str[j]:
                     filter_out.append(string[i])
    return (filter_out)

myfilter(["new","old","replace"],["replace","new"])     
```

```python
input_list=['aa','abc','abcd','abcdef']

def longestWord(input_list):
    len_of_list=[len(i) for i in input_list]
    len_of_list.sort(reverse=True)
    max_length=len_of_list[0]
    for i in input_list:
        if len(i)==max_length:
            return i
longestWord(input_list)
```

```python

class AreaofTriangle:
    def __init__(self, a, b,c):
        self.a = a
        self.b = b
        self.c = c

    def Area(self):
        self.s=(self.a+self.b+self.c)/2
        return ('%0.2f' %((self.s*(self.s-self.a)*(self.s-self.b)*(self.s-self.c)) ** 0.5))
        

ob1 = AreaofTriangle(5,6,7)
ob1.Area()
```

```python
input_list=['aa','abc','abcd','abcdef','w','wxfrg']

def filter_long_words(input_list,n):
    long_word_list=[]
    len_of_list=[len(i) for i in input_list]
    for i in input_list:
        if len(i)>n:
            long_word_list.append(i)
    return long_word_list
            
filter_long_words(input_list,2)
```

```python
vowel_list=['a','e','i','o','u']

def chk_vowel(instr):
    value=False
    for letter in vowel_list:
        if instr.lower()==letter:
            value=True
    return value

chk_vowel('o')
```
