```python
#Approach 1

input_array=[3, 5, 7, 2, 8, 10, 11, 65, 72, 81, 99, 100, 150]
k=3
n=len(input_array)
def moving_average(n,k,input_array):
    out_mv_avg=[]
    for i in range(0,n-k+1):
        s=0
        for j in range(i,i+k):
            s=s+input_array[j]
        s=s/k
        out_mv_avg.append(s)
    return out_mv_avg
moving_average(n,k,input_array)  
    
```

```python
#Approach 2
import numpy as np
from numpy import convolve
input_array=[3, 5, 7, 2, 8, 10, 11, 65, 72, 81, 99, 100, 150]
def movingaverage (n, k):
    divisor = np.repeat(1.0, k)/k
    result = np.convolve(n, divisor, 'valid')
    return result
movingaverage (input_array, 3)
```
