```python
subjects=["Americans","Indians"]
verbs=["play","watch"]
objects=["Baseball.","Cricket."]

def new(x,y,z):
    return [(x+" "+y+" "+z) for x in subjects for y in verbs for z in objects]

for i in new(subjects,verbs,objects):
    print(i)
    
```
