```python
import numpy as np

Max = np.array([39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25])
Min = np.array([21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18])

import matplotlib.pyplot as plt
months = np.arange(12)
```

```python
from scipy import optimize
def yearly_temps(times, avg, ampl, time_offset):
    #print(ampl)
    return (avg
            + ampl * np.cos((times + time_offset) * 2 * np.pi / times.max()))

res_max, cov_max = optimize.curve_fit(yearly_temps, months,
                                      Max, [20, 10, 0])


#print(res_max)

res_min, cov_min = optimize.curve_fit(yearly_temps, months,
                                      Min, [-40, 20, 0])
```

```python
days = np.linspace(0, 12, num=365)


plt.figure()
plt.plot(months, Max, 'ro')
plt.plot(days, yearly_temps(days, *res_max), 'r-')
plt.plot(months, Min, 'bo')
plt.plot(days, yearly_temps(days, *res_min), 'b-')
plt.xlabel('Month')
plt.ylabel('Temperature ($^\circ$C)')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url= "https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv"
titanic = pd.read_csv(url, skipfooter=1, engine='python')

lab = titanic["sex"].value_counts().keys().tolist()
#values
val = titanic["sex"].value_counts().values.tolist()
from matplotlib import pyplot as plt
title = "pie chart"
plt.pie(val, labels=lab, autopct = '%.1f%%')
plt.axis('equal')
plt.title('Male/Female proportion')
plt.tight_layout()
plt.show()
 

colors = {'male':'orange', 'female':'blue'}
fig2, ax2 = plt.subplots()
ax2.scatter(titanic['age'], titanic['fare'], alpha=0.5, c=titanic['sex'].apply(lambda x: colors[x]))
ax2.set_xlabel('Fare')
ax2.set_ylabel('Age')
ax2.set_title('Fare paid and the Age')
```
